**First dataset**

```python
ev_states=pd.read_csv("EV_India.csv")
```

**Deleating rows with No data**

```python
rows = ev_states.index[[1,17,18,30]]

ev_states.drop(rows, inplace=True)

ev_states.drop(['Sr. No.'], axis = 1, inplace = True)
```

```python
# Convert string numbers to integers
ev_states['Total Electric Vehicle'] = ev_states['Total Electric Vehicle'].apply(lambda x: int(str(x).replace(",", "")))
ev_states['Total Non-Electric Vehicle'] = ev_states['Total Non-Electric Vehicle'].apply(lambda x: int(str(x).replace(",", "")))
ev_states['Total'] = ev_states['Total'].apply(lambda x: int(str(x).replace(",", "")))

```

**Identifying best clusters**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X = ev_customers[features[:-1]]

# Standardize the feature values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters using the elbow method
inertia = []
k_values = range(1, 11)
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve to identify the optimal number of clusters
plt.plot(k_values, inertia, 'o-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.show()


```

**Performing K-means according to best clustering**

```python
def perform_k_means_and_pairplot(features,k=3):
    features = features

    scaler = StandardScaler()
    ev_customers_scaled = scaler.fit_transform(ev_customers[features])

    k = k
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(ev_customers_scaled)

    ev_customers['Cluster'] = kmeans.labels_

    sns.pairplot(ev_customers, vars=features, hue='Cluster', palette='viridis')
    plt.suptitle(f'Feature Visualization with Clusters for {features}', y=1.02)
    plt.show()
```

**Identifying best clusters according to Heirarchichal Clustering**

```python
X = ev_brand.iloc[:,:].values


# Using the dendrogram to find the optimal number of clusters
plt.figure(figsize=(15,7))
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
plt.title('Dendrogram')

plt.show()
```

**Identifying best clusters via elbow graph**

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

**K-Means with best cluster number**

```python
def perform_k_means_and_pairplot(features, k=3):
    features = features
    scaler = StandardScaler()
    ev_brand_scaled = scaler.fit_transform(ev_brand[features])


    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(ev_brand_scaled)

    ev_brand['Cluster'] = kmeans.labels_

    sns.pairplot(ev_brand, vars=features, hue='Cluster', palette='viridis')
    plt.suptitle(f'Feature Visualization with Clusters for {features}', y=1.02)
    plt.show()
```
